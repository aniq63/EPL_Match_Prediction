```python
# ============================================================
# CELL 2 — Imports
# ============================================================

import pandas as pd
import numpy as np
import soccerdata as sd
import warnings
import matplotlib.pyplot as plt

from datetime import datetime
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.base import BaseEstimator
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import RandomizedSearchCV, TimeSeriesSplit
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    ConfusionMatrixDisplay
)
from scipy.stats import randint, uniform

warnings.filterwarnings("ignore")
pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:.3f}".format)

print("All imports OK")


# ============================================================
# CELL 3 — Row tracker (run once, used throughout)
# ============================================================
#
# Every pipeline step calls track() before and after.
# At the end, print_audit() shows exactly where rows were lost,
# how many, and why — so you always know you're not accidentally
# training on a shrinking subset.
# ============================================================

class RowTracker:
    def __init__(self):
        self.log = []          # list of dicts
        self._current = None   # snapshot taken at track_before()

    def before(self, step_name: str, df: pd.DataFrame):
        """Call immediately before a pipeline step."""
        self._current = {"step": step_name, "before": len(df)}

    def after(self, df: pd.DataFrame, note: str = ""):
        """Call immediately after the step with the result df."""
        assert self._current is not None, "Call .before() first"
        entry = {
            **self._current,
            "after":   len(df),
            "dropped": self._current["before"] - len(df),
            "note":    note,
        }
        self.log.append(entry)
        dropped = entry["dropped"]
        flag = "  ⚠️  ROWS DROPPED" if dropped > 0 else ""
        print(f"  [{entry['step']}]  {entry['before']} → {entry['after']}"
              f"  (dropped {dropped}){flag}")
        self._current = None

    def print_audit(self):
        print("\n" + "=" * 65)
        print("  PIPELINE ROW AUDIT")
        print("=" * 65)
        print(f"  {'Step':<38}  {'Before':>7}  {'After':>7}  {'Lost':>6}")
        print(f"  {'-'*60}")
        for e in self.log:
            flag = " !" if e["dropped"] > 0 else ""
            print(f"  {e['step']:<38}  {e['before']:>7}  {e['after']:>7}"
                  f"  {e['dropped']:>6}{flag}")
            if e["note"]:
                print(f"    └─ {e['note']}")
        total_in  = self.log[0]["before"] if self.log else 0
        total_out = self.log[-1]["after"]  if self.log else 0
        print(f"  {'-'*60}")
        print(f"  {'TOTAL':<38}  {total_in:>7}  {total_out:>7}"
              f"  {total_in - total_out:>6}")
        pct = (total_out / total_in * 100) if total_in > 0 else 0
        print(f"\n  Rows retained: {pct:.1f}%")
        print("=" * 65)

tracker = RowTracker()
print("RowTracker ready — will audit every pipeline step")


# ============================================================
# CELL 4 — Season auto-detection
# ============================================================

def get_seasons(base_year: int = 2023) -> list[str]:
    now = datetime.now()
    current_start = now.year if now.month >= 8 else now.year - 1
    return [f"{y}/{y+1}" for y in range(base_year, current_start + 1)]

SEASONS = get_seasons(base_year=2023)
print(f"Seasons to fetch: {SEASONS}")


# ============================================================
# CELL 5 — Load raw data from Understat
# ============================================================

def load_understat_data(seasons: list[str],
                        league: str = "ENG-Premier League") -> pd.DataFrame:
    dfs = []
    for season in seasons:
        print(f"  Fetching {season} ...", end=" ")
        understat = sd.Understat(leagues=league, seasons=season)
        df_s = understat.read_team_match_stats()
        dfs.append(df_s)
        print(f"got {len(df_s)} rows")
    df = pd.concat(dfs, ignore_index=True)
    print(f"\nTotal raw rows: {df.shape[0]}  |  Columns: {df.shape[1]}")
    return df

print("Loading data from Understat...")
raw_df = load_understat_data(SEASONS)
print("\nColumns available:")
print(list(raw_df.columns))


# ============================================================
# CELL 6 — Drop useless columns
# ============================================================

COLS_TO_DROP = [
    "league_id", "season_id", "game_id",
    "home_team_id", "away_team_id",
    "home_team_code", "away_team_code",
    "home_expected_points", "away_expected_points",
    "home_np_xg_difference", "away_np_xg_difference",
    # np_xg ~0.95 correlated with xg — adds noise not signal
    "home_np_xg", "away_np_xg",
]

def drop_useless_columns(df: pd.DataFrame) -> pd.DataFrame:
    existing = [c for c in COLS_TO_DROP if c in df.columns]
    df = df.drop(columns=existing)
    print(f"Dropped {len(existing)} columns: {existing}")
    return df

tracker.before("drop_useless_columns", raw_df)
df = drop_useless_columns(raw_df.copy())
tracker.after(df, "column drop only — no rows should be lost here")
print(f"Shape after drop: {df.shape}")


# ============================================================
# CELL 7 — Feature engineering
# ============================================================

def feature_engineering(df: pd.DataFrame) -> pd.DataFrame:
    df["date"] = pd.to_datetime(df["date"])
    df.sort_values("date", inplace=True)
    df.reset_index(drop=True, inplace=True)

    df["result"] = df.apply(
        lambda r: "Win"  if r["home_goals"] > r["away_goals"]
        else      "Draw" if r["home_goals"] == r["away_goals"]
        else      "Lose",
        axis=1
    )

    df["home_goals_conceded"] = df["away_goals"]
    df["away_goals_conceded"] = df["home_goals"]

    print(f"Result distribution:\n{df['result'].value_counts()}")
    return df

tracker.before("feature_engineering", df)
df = feature_engineering(df)
tracker.after(df, "no rows should be dropped here")
print(f"Shape after feature engineering: {df.shape}")


# ============================================================
# CELL 8 — Rolling averages (last 5 matches, per team)
# ============================================================
#
# KEY DESIGN: unified team timeline
#
# We stack every match into two rows — one per team —
# then compute rolling on that combined timeline.
# This means each team's last-5 average uses ALL their
# recent matches regardless of home/away venue.
#
# Previous versions grouped by home_team only — away
# performances were silently ignored for every team.
#
# Merge strategy: inner join on (date, home_team) and
# (date, away_team). An inner join here is correct because
# any match with no rolling data yet (early season, < 5 games)
# should be excluded — we can't make a valid prediction
# without a full 5-game window.
# ============================================================

STAT_COLS = [
    "goals", "goals_conceded",
    "xg", "ppda", "deep_completions",
]

def build_team_timeline(df: pd.DataFrame) -> pd.DataFrame:
    home = df[[
        "date", "home_team",
        "home_goals", "home_goals_conceded",
        "home_xg", "home_ppda", "home_deep_completions",
    ]].rename(columns={
        "home_team":             "team",
        "home_goals":            "goals",
        "home_goals_conceded":   "goals_conceded",
        "home_xg":               "xg",
        "home_ppda":             "ppda",
        "home_deep_completions": "deep_completions",
    })
    home["venue"] = "home"

    away = df[[
        "date", "away_team",
        "away_goals", "away_goals_conceded",
        "away_xg", "away_ppda", "away_deep_completions",
    ]].rename(columns={
        "away_team":             "team",
        "away_goals":            "goals",
        "away_goals_conceded":   "goals_conceded",
        "away_xg":               "xg",
        "away_ppda":             "ppda",
        "away_deep_completions": "deep_completions",
    })
    away["venue"] = "away"

    tl = pd.concat([home, away], ignore_index=True)
    tl.sort_values(["team", "date"], inplace=True)
    return tl

def add_rolling_to_timeline(tl: pd.DataFrame,
                             cols: list, window: int = 5) -> pd.DataFrame:
    grp = tl.groupby("team")
    for col in cols:
        tl[f"{col}_avg_last5"] = (
            grp[col]
            .rolling(window, closed="left")
            .mean()
            .reset_index(0, drop=True)
        )
    avg_cols = [f"{c}_avg_last5" for c in cols]
    rows_before = len(tl)
    tl.dropna(subset=avg_cols, inplace=True)
    print(f"  Timeline: {rows_before} team-match rows → {len(tl)} "
          f"after dropping < {window}-game window "
          f"(dropped {rows_before - len(tl)})")
    return tl

def merge_rolling_back(df: pd.DataFrame,
                       tl: pd.DataFrame) -> pd.DataFrame:
    """
    Merge strategy: LEFT join, then drop NaN.
    Using LEFT (not inner) so we can explicitly see and count
    which matches got dropped — inner would hide the losses.
    """
    avg_cols = [f"{c}_avg_last5" for c in STAT_COLS]
    keep = ["date", "team"] + avg_cols

    home_avgs = tl[tl["venue"] == "home"][keep].rename(
        columns={"team": "home_team",
                 **{c: f"home_{c}" for c in avg_cols}}
    )
    away_avgs = tl[tl["venue"] == "away"][keep].rename(
        columns={"team": "away_team",
                 **{c: f"away_{c}" for c in avg_cols}}
    )

    n0 = len(df)
    df = df.merge(home_avgs, on=["date", "home_team"], how="left")
    n1 = len(df)
    df = df.merge(away_avgs, on=["date", "away_team"], how="left")
    n2 = len(df)

    # LEFT merges should never change row count
    assert n0 == n1 == n2, (
        f"Merge changed row count: {n0} → {n1} → {n2}. "
        "Check for duplicate keys in rolling table."
    )

    # Drop rows where rolling window wasn't full yet
    drop_cols = [f"home_{c}" for c in avg_cols] + \
                [f"away_{c}" for c in avg_cols]
    rows_before = len(df)
    df.dropna(subset=drop_cols, inplace=True)
    print(f"  Matches dropped for insufficient rolling window: "
          f"{rows_before - len(df)}")
    return df

tracker.before("rolling_averages", df)
print("Building team timeline and computing rolling averages...")
team_tl = build_team_timeline(df)
team_tl = add_rolling_to_timeline(team_tl, STAT_COLS)
df      = merge_rolling_back(df, team_tl)
tracker.after(df, "rows dropped = matches with < 5-game history (expected, early season)")
print(f"Shape after rolling averages: {df.shape}")


# ============================================================
# CELL 9 — Points collected (last 5 matches, per team)
# ============================================================
#
# Merge strategy: LEFT join + explicit NaN drop + assertion.
# The left join row count must stay identical to input —
# if it changes, there are duplicate keys in the points table.
# ============================================================

def add_points_last5(df: pd.DataFrame) -> pd.DataFrame:
    home_pts = df[["date", "home_team", "home_points"]].rename(
        columns={"home_team": "team", "home_points": "points"}
    )
    away_pts = df[["date", "away_team", "away_points"]].rename(
        columns={"away_team": "team", "away_points": "points"}
    )

    tm = pd.concat([home_pts, away_pts], ignore_index=True)
    tm.sort_values(["team", "date"], inplace=True)

    tm["pts_last5"] = (
        tm.groupby("team")["points"]
          .rolling(5, closed="left")
          .sum()
          .reset_index(0, drop=True)
    )
    tm.dropna(subset=["pts_last5"], inplace=True)
    pts_cols = ["date", "team", "pts_last5"]

    n0 = len(df)

    df = df.merge(
        tm[pts_cols], left_on=["date", "home_team"],
        right_on=["date", "team"], how="left"
    ).rename(columns={"pts_last5": "home_points_last5"}).drop(columns="team")

    assert len(df) == n0, \
        f"Home points merge changed row count {n0} → {len(df)}. Duplicate keys?"

    df = df.merge(
        tm[pts_cols], left_on=["date", "away_team"],
        right_on=["date", "team"], how="left"
    ).rename(columns={"pts_last5": "away_points_last5"}).drop(columns="team")

    assert len(df) == n0, \
        f"Away points merge changed row count {n0} → {len(df)}. Duplicate keys?"

    rows_before = len(df)
    df.dropna(subset=["home_points_last5", "away_points_last5"], inplace=True)
    print(f"  Points NaN drop: {rows_before - len(df)} rows "
          f"(matches before 5-game points window filled)")
    return df

tracker.before("add_points_last5", df)
df = add_points_last5(df)
tracker.after(df, "NaN rows = matches before 5-game points window")
print(f"Shape after points last5: {df.shape}")


# ============================================================
# CELL 10 — Venue-specific W/D/L form (last 5 home / last 5 away)
# ============================================================
#
# Bug note: ag["a_draw"] and ag["a_loss"] must reference ag,
# NOT hg. After sort_values the two DataFrames have different
# row orders — cross-referencing silently computes wrong labels.
#
# Merge strategy: LEFT + assertion on row count.
# NaN rows (< 5 venue games) are tracked and dropped explicitly.
# ============================================================

def add_venue_form_rolling(df: pd.DataFrame) -> pd.DataFrame:
    n0 = len(df)

    # ── Home-venue form ──────────────────────────────────────
    hg = df[["date", "home_team", "home_goals", "away_goals"]].copy()
    hg.sort_values(["home_team", "date"], inplace=True)

    hg["h_win"]  = (hg["home_goals"] > hg["away_goals"]).astype(int)
    hg["h_draw"] = (hg["home_goals"] == hg["away_goals"]).astype(int)
    hg["h_loss"] = (hg["home_goals"] < hg["away_goals"]).astype(int)

    grp_h = hg.groupby("home_team")
    hg["hw5"] = grp_h["h_win"].rolling(5,  closed="left").sum().reset_index(0, drop=True)
    hg["hd5"] = grp_h["h_draw"].rolling(5, closed="left").sum().reset_index(0, drop=True)
    hg["hl5"] = grp_h["h_loss"].rolling(5, closed="left").sum().reset_index(0, drop=True)
    hg["hwr"] = hg["hw5"] / 5.0

    df = df.merge(
        hg[["date", "home_team", "hw5", "hd5", "hl5", "hwr"]],
        on=["date", "home_team"], how="left"
    )
    assert len(df) == n0, \
        f"Home venue merge changed row count {n0} → {len(df)}."

    # ── Away-venue form ──────────────────────────────────────
    # IMPORTANT: all three comparisons use ag, never hg.
    # hg is sorted by home_team+date; ag by away_team+date.
    # Cross-referencing gives silently wrong results.
    ag = df[["date", "away_team", "away_goals", "home_goals"]].copy()
    ag.sort_values(["away_team", "date"], inplace=True)

    ag["a_win"]  = (ag["away_goals"] > ag["home_goals"]).astype(int)
    ag["a_draw"] = (ag["away_goals"] == ag["home_goals"]).astype(int)  # ag, not hg
    ag["a_loss"] = (ag["away_goals"] < ag["home_goals"]).astype(int)   # ag, not hg

    grp_a = ag.groupby("away_team")
    ag["aw5"] = grp_a["a_win"].rolling(5,  closed="left").sum().reset_index(0, drop=True)
    ag["ad5"] = grp_a["a_draw"].rolling(5, closed="left").sum().reset_index(0, drop=True)
    ag["al5"] = grp_a["a_loss"].rolling(5, closed="left").sum().reset_index(0, drop=True)
    ag["awr"] = ag["aw5"] / 5.0

    df = df.merge(
        ag[["date", "away_team", "aw5", "ad5", "al5", "awr"]],
        on=["date", "away_team"], how="left"
    )
    assert len(df) == n0, \
        f"Away venue merge changed row count {n0} → {len(df)}."

    df.rename(columns={
        "hw5": "home_team_home_wins_last5",
        "hd5": "home_team_home_draws_last5",
        "hl5": "home_team_home_losses_last5",
        "aw5": "away_team_away_wins_last5",
        "ad5": "away_team_away_draws_last5",
        "al5": "away_team_away_losses_last5",
    }, inplace=True)

    df["home_venue_advantage"] = df["hwr"] - df["awr"]
    df.drop(columns=["hwr", "awr"], inplace=True)

    venue_cols = ["home_team_home_wins_last5", "away_team_away_wins_last5"]
    rows_before = len(df)
    df.dropna(subset=venue_cols, inplace=True)
    print(f"  Venue NaN drop: {rows_before - len(df)} rows "
          f"(teams with < 5 home or away games)")
    return df

tracker.before("venue_form_rolling", df)
df = add_venue_form_rolling(df)
tracker.after(df, "NaN rows = teams with < 5 venue-specific games")
print(f"Shape after venue form rolling: {df.shape}")


# ============================================================
# CELL 11 — Drop raw input columns
# ============================================================

RAW_STAT_COLS = [
    "home_goals", "away_goals",
    "home_goals_conceded", "away_goals_conceded",
    "home_xg", "away_xg",
    "home_ppda", "away_ppda",
    "home_deep_completions", "away_deep_completions",
    "home_points", "away_points",
]

tracker.before("drop_raw_columns", df)
df.drop(columns=[c for c in RAW_STAT_COLS if c in df.columns], inplace=True)
tracker.after(df, "column drop only — no rows should be lost")
print(f"Shape after dropping raw columns: {df.shape}")


# ============================================================
# CELL 12 — Derived difference features
# ============================================================

def add_derived_features(df: pd.DataFrame) -> pd.DataFrame:
    df["points_diff_last5"]   = df["home_points_last5"]               - df["away_points_last5"]
    df["goal_diff_avg5"]      = df["home_goals_avg_last5"]            - df["away_goals_avg_last5"]
    df["xg_diff_avg5"]        = df["home_xg_avg_last5"]               - df["away_xg_avg_last5"]
    df["x_defense_diff"]      = df["away_goals_conceded_avg_last5"]   - df["home_goals_conceded_avg_last5"]
    df["ppda_diff_avg5"]      = df["home_ppda_avg_last5"]             - df["away_ppda_avg_last5"]
    df["deep_comp_diff_avg5"] = df["home_deep_completions_avg_last5"] - df["away_deep_completions_avg_last5"]
    df["venue_wins_diff"]     = df["home_team_home_wins_last5"]       - df["away_team_away_wins_last5"]
    df["home_advantage"]      = 1
    return df

tracker.before("add_derived_features", df)
df = add_derived_features(df)
df.sort_values("date", inplace=True)
df.reset_index(drop=True, inplace=True)
tracker.after(df, "pure computation — no rows should be lost")

print(f"Final dataset shape: {df.shape}")
print(f"\nFull column list:\n{list(df.columns)}")

# Print the full audit now that all feature engineering is done
tracker.print_audit()


# ============================================================
# CELL 13 — Train / Test split (temporal)
# ============================================================

SPLIT_DATE = "2026-03-01"

train_df = df[df["date"] < SPLIT_DATE].copy()
test_df  = df[df["date"] >= SPLIT_DATE].copy()

print(f"Training rows : {len(train_df)}  ({train_df['date'].min().date()} → {train_df['date'].max().date()})")
print(f"Test rows     : {len(test_df)}  ({test_df['date'].min().date()} → {test_df['date'].max().date()})")
print(f"\nTrain result distribution:\n{train_df['result'].value_counts()}")
print(f"\nTest result distribution:\n{test_df['result'].value_counts()}")


# ============================================================
# CELL 14 — Prepare X / y for model
# ============================================================

INPUT_FEATURES = [
    # ── Per-team rolling averages (home+away combined) ───────
    "home_goals_avg_last5",
    "away_goals_avg_last5",
    "home_goals_conceded_avg_last5",
    "away_goals_conceded_avg_last5",
    "home_xg_avg_last5",
    "away_xg_avg_last5",
    "home_ppda_avg_last5",
    "away_ppda_avg_last5",
    "home_deep_completions_avg_last5",
    "away_deep_completions_avg_last5",
    # ── Points collected last 5 ──────────────────────────────
    "home_points_last5",
    "away_points_last5",
    # ── Venue-specific W/D/L counts ──────────────────────────
    "home_team_home_wins_last5",
    "home_team_home_draws_last5",
    "home_team_home_losses_last5",
    "away_team_away_wins_last5",
    "away_team_away_draws_last5",
    "away_team_away_losses_last5",
    # ── Difference features ──────────────────────────────────
    "points_diff_last5",
    "goal_diff_avg5",
    "xg_diff_avg5",
    "x_defense_diff",
    "ppda_diff_avg5",
    "deep_comp_diff_avg5",
    "venue_wins_diff",
    "home_venue_advantage",
    # ── Contextual ───────────────────────────────────────────
    "home_advantage",
]

le = LabelEncoder()
le.fit(["Win", "Draw", "Lose"])

X_train = train_df[INPUT_FEATURES]
y_train = le.transform(train_df["result"])

X_test  = test_df[INPUT_FEATURES]
y_test  = le.transform(test_df["result"])

print(f"X_train shape : {X_train.shape}")
print(f"X_test  shape : {X_test.shape}")
print(f"Total features: {len(INPUT_FEATURES)}")
print(f"Classes       : {le.classes_}")


# ============================================================
# CELL 15 — RandomizedSearchCV (hyperparameter search)
# ============================================================
#
# WHAT IS RandomizedSearchCV?
#   Instead of trying every combination (GridSearch),
#   RandomizedSearch samples n_iter random combinations
#   from the parameter distributions. Much faster when the
#   search space is large.
#
# CROSS-VALIDATION STRATEGY — TimeSeriesSplit:
#   We use TimeSeriesSplit(n_splits=5) instead of the default
#   KFold. Why? Our data is time-ordered — a random fold would
#   let future matches train the model on past matches it's
#   being evaluated on (data leakage). TimeSeriesSplit always
#   trains on earlier data and validates on later data.
#
#   Split layout (5 folds):
#     Fold 1: train[0:20%]  → val[20:40%]
#     Fold 2: train[0:40%]  → val[40:60%]
#     Fold 3: train[0:60%]  → val[60:80%]
#     Fold 4: train[0:80%]  → val[80:100%]
#
# PARAMETER SPACE:
#   DecisionTree (base estimator):
#     max_depth        — how deep each tree splits
#     min_samples_leaf — min samples at leaf, controls overfitting
#     min_samples_split— min samples to attempt a split
#     max_features     — features considered at each split
#
#   AdaBoost (wrapper):
#     n_estimators     — number of boosting rounds
#     learning_rate    — shrinkage per round
#
# NOTE: NO class_weight on the DecisionTree.
#   AdaBoost does its own sample reweighting each round.
#   Adding class_weight fights that mechanism directly.
#
# n_iter=50 tries 50 random combinations.
# Scoring="accuracy" — optimises for overall match accuracy.
# n_jobs=-1 uses all CPU cores.
# ============================================================

print("Setting up RandomizedSearchCV...")
print(f"Training on {len(X_train)} rows  |  {len(INPUT_FEATURES)} features")
print("Cross-validation: TimeSeriesSplit(n_splits=5) — no future leakage\n")

# ── Parameter distributions ──────────────────────────────────
param_dist = {
    # DecisionTree base estimator params (prefixed with estimator__)
    "estimator__max_depth":         randint(1, 8),        # 1 to 7
    "estimator__min_samples_leaf":  randint(1, 20),       # 1 to 19
    "estimator__min_samples_split": randint(2, 20),       # 2 to 19
    "estimator__max_features":      ["sqrt", "log2", None],

    # AdaBoost params
    "n_estimators":  randint(100, 600),       # 100 to 599
    "learning_rate": uniform(0.01, 0.49),     # 0.01 to 0.50
}

# ── Base pipeline for search ──────────────────────────────────
search_base = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(random_state=42),
    algorithm="SAMME",
    random_state=42
)

# ── Time-series cross-validation splitter ────────────────────
tscv = TimeSeriesSplit(n_splits=5)

# ── RandomizedSearchCV ────────────────────────────────────────
random_search = RandomizedSearchCV(
    estimator=search_base,
    param_distributions=param_dist,
    n_iter=50,              # number of random combinations to try
    scoring="accuracy",     # optimise for match prediction accuracy
    cv=tscv,                # temporal splits — no future leakage
    n_jobs=-1,              # use all CPU cores
    verbose=2,              # print progress per fold
    random_state=42,
    refit=True,             # refit best params on full X_train
    return_train_score=True
)

print("Starting search (50 iterations × 5 folds = 250 fits)...")
print("This will take 3–8 minutes depending on your machine.\n")
random_search.fit(X_train, y_train)

# ── Print results ─────────────────────────────────────────────
print("\n" + "=" * 55)
print("  RANDOMIZED SEARCH RESULTS")
print("=" * 55)
print(f"  Best CV accuracy : {random_search.best_score_:.4f}"
      f"  ({random_search.best_score_*100:.2f}%)")
print(f"\n  Best parameters found:")

bp = random_search.best_params_
print(f"    AdaBoost:")
print(f"      n_estimators  : {bp['n_estimators']}")
print(f"      learning_rate : {bp['learning_rate']:.4f}")
print(f"    DecisionTree (base estimator):")
print(f"      max_depth         : {bp['estimator__max_depth']}")
print(f"      min_samples_leaf  : {bp['estimator__min_samples_leaf']}")
print(f"      min_samples_split : {bp['estimator__min_samples_split']}")
print(f"      max_features      : {bp['estimator__max_features']}")
print("=" * 55)

# ── Top 10 combinations from the search ──────────────────────
cv_results = pd.DataFrame(random_search.cv_results_)
top10 = (
    cv_results[["mean_test_score", "std_test_score", "params"]]
    .sort_values("mean_test_score", ascending=False)
    .head(10)
    .reset_index(drop=True)
)
print("\nTop 10 parameter combinations:")
for i, row in top10.iterrows():
    print(f"  {i+1:2d}.  acc={row['mean_test_score']:.4f} "
          f"±{row['std_test_score']:.4f}  "
          f"n_est={row['params']['n_estimators']}  "
          f"lr={row['params']['learning_rate']:.3f}  "
          f"depth={row['params']['estimator__max_depth']}  "
          f"leaf={row['params']['estimator__min_samples_leaf']}")


# ============================================================
# CELL 16 — Build final tuned model from best parameters
# ============================================================
#
# We take the best parameters found by RandomizedSearchCV
# and explicitly construct both the DecisionTree base estimator
# and the AdaBoost wrapper — so every parameter is visible
# and documented, not hidden inside a black-box .best_estimator_.
#
# Training on the FULL X_train (not just one CV fold).
# ============================================================

print("\nBuilding final tuned model from best parameters...")

# Extract best params cleanly
best_max_depth          = bp["estimator__max_depth"]
best_min_samples_leaf   = bp["estimator__min_samples_leaf"]
best_min_samples_split  = bp["estimator__min_samples_split"]
best_max_features       = bp["estimator__max_features"]
best_n_estimators       = bp["n_estimators"]
best_learning_rate      = bp["learning_rate"]

# ── Tuned DecisionTree base estimator ────────────────────────
tuned_tree = DecisionTreeClassifier(
    max_depth         = best_max_depth,
    min_samples_leaf  = best_min_samples_leaf,
    min_samples_split = best_min_samples_split,
    max_features      = best_max_features,
    # NO class_weight — AdaBoost handles sample weighting itself
    random_state=42
)

# ── Tuned AdaBoost ────────────────────────────────────────────
model = AdaBoostClassifier(
    estimator     = tuned_tree,
    n_estimators  = best_n_estimators,
    learning_rate = best_learning_rate,
    algorithm     = "SAMME",
    random_state  = 42
)

print(f"Training final AdaBoost on {len(X_train)} rows...")
model.fit(X_train, y_train)
print("Training complete.")


# ============================================================
# CELL 17 — Evaluate tuned model on test set
# ============================================================

y_pred       = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)

acc    = accuracy_score(y_test, y_pred)
cv_acc = random_search.best_score_

print(f"\nCV best accuracy  (train)  : {cv_acc:.4f}  ({cv_acc*100:.2f}%)")
print(f"Test accuracy     (unseen) : {acc:.4f}  ({acc*100:.2f}%)")

gap = cv_acc - acc
if abs(gap) < 0.03:
    print("  Gap < 3% — model is generalising well")
elif gap > 0.05:
    print(f"  Gap = {gap:.2f} — possible overfitting, consider more regularisation")
else:
    print(f"  Gap = {gap:.2f} — acceptable")

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_, digits=4))


# ============================================================
# CELL 18 — Confusion matrix + feature importance plot
# ============================================================

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

cm   = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=le.classes_)
disp.plot(ax=axes[0], colorbar=False, cmap="Blues")
axes[0].set_title("Confusion Matrix (Tuned AdaBoost)", fontsize=13, fontweight="bold")

feat_imp        = pd.Series(model.feature_importances_, index=INPUT_FEATURES)
feat_imp_sorted = feat_imp.sort_values(ascending=True).tail(20)
feat_imp_sorted.plot(kind="barh", ax=axes[1], color="steelblue", edgecolor="white")
axes[1].set_title("Top 20 Feature Importances", fontsize=13, fontweight="bold")
axes[1].set_xlabel("Importance")
axes[1].tick_params(axis="y", labelsize=9)

plt.tight_layout()
plt.savefig("evaluation_plots.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved as evaluation_plots.png")

print("\nLowest importance features (consider dropping next round):")
print(feat_imp.sort_values().head(8).to_string())


# ============================================================
# CELL 19 — Predictor class
# ============================================================

class EPLPredictor:

    def __init__(self, historical_df: pd.DataFrame,
                 model: BaseEstimator,
                 label_encoder: LabelEncoder,
                 feature_cols: list[str]):
        keep = [
            "date", "home_team", "away_team",
            "home_goals", "away_goals",
            "home_xg", "away_xg",
            "home_ppda", "away_ppda",
            "home_deep_completions", "away_deep_completions",
            "home_points", "away_points",
        ]
        self.raw_df = historical_df[keep].copy()
        self.raw_df["date"] = pd.to_datetime(self.raw_df["date"])
        self.raw_df["home_goals_conceded"] = self.raw_df["away_goals"]
        self.raw_df["away_goals_conceded"] = self.raw_df["home_goals"]
        self.model        = model
        self.le           = label_encoder
        self.feature_cols = feature_cols

    def _last5_combined(self, team: str, before: pd.Timestamp) -> pd.DataFrame:
        df   = self.raw_df
        mask = (
            ((df["home_team"] == team) | (df["away_team"] == team)) &
            (df["date"] < before)
        )
        last5 = df[mask].sort_values("date").tail(5)
        if len(last5) < 5:
            raise ValueError(
                f"Not enough history for '{team}' before {before.date()} "
                f"(found {len(last5)}/5). Try a later match date."
            )
        return last5

    def _last5_home(self, team: str, before: pd.Timestamp) -> pd.DataFrame:
        df    = self.raw_df
        mask  = (df["home_team"] == team) & (df["date"] < before)
        last5 = df[mask].sort_values("date").tail(5)
        if len(last5) < 5:
            raise ValueError(
                f"Not enough HOME history for '{team}' before {before.date()} "
                f"(found {len(last5)}/5). Try a later match date."
            )
        return last5

    def _last5_away(self, team: str, before: pd.Timestamp) -> pd.DataFrame:
        df    = self.raw_df
        mask  = (df["away_team"] == team) & (df["date"] < before)
        last5 = df[mask].sort_values("date").tail(5)
        if len(last5) < 5:
            raise ValueError(
                f"Not enough AWAY history for '{team}' before {before.date()} "
                f"(found {len(last5)}/5). Try a later match date."
            )
        return last5

    def _rolling_form(self, matches: pd.DataFrame, team: str) -> dict:
        goals, conceded, xg, ppda, deep = [], [], [], [], []
        for _, r in matches.iterrows():
            side = "home" if r["home_team"] == team else "away"
            opp  = "away" if side == "home" else "home"
            goals.append(r[f"{side}_goals"])
            conceded.append(r[f"{opp}_goals"])
            xg.append(r[f"{side}_xg"])
            ppda.append(r[f"{side}_ppda"])
            deep.append(r[f"{side}_deep_completions"])
        return {
            "goals":    np.mean(goals),
            "conceded": np.mean(conceded),
            "xg":       np.mean(xg),
            "ppda":     np.mean(ppda),
            "deep":     np.mean(deep),
        }

    def _points_last5(self, matches: pd.DataFrame, team: str) -> float:
        total = 0.0
        for _, r in matches.iterrows():
            col    = "home_points" if r["home_team"] == team else "away_points"
            total += float(r[col])
        return total

    def _venue_wdl(self, matches: pd.DataFrame, side: str) -> dict:
        # side and opp always reference the same DataFrame (matches)
        # — no cross-DataFrame comparison possible here
        opp = "away" if side == "home" else "home"
        wins, draws, losses = 0, 0, 0
        for _, r in matches.iterrows():
            gf = r[f"{side}_goals"]
            ga = r[f"{opp}_goals"]
            if gf > ga:    wins   += 1
            elif gf == ga: draws  += 1
            else:          losses += 1
        return {"wins": wins, "draws": draws, "losses": losses,
                "win_rate": wins / 5.0}

    def predict(self, home_team: str, away_team: str,
                match_date: str) -> dict:
        md = pd.to_datetime(match_date)

        h5  = self._last5_combined(home_team, md)
        a5  = self._last5_combined(away_team, md)
        hrf = self._rolling_form(h5, home_team)
        arf = self._rolling_form(a5, away_team)
        hpt = self._points_last5(h5, home_team)
        apt = self._points_last5(a5, away_team)

        h5_home = self._last5_home(home_team, md)
        a5_away = self._last5_away(away_team, md)
        hvf     = self._venue_wdl(h5_home, "home")
        avf     = self._venue_wdl(a5_away, "away")

        row = {
            "home_goals_avg_last5":              hrf["goals"],
            "away_goals_avg_last5":              arf["goals"],
            "home_goals_conceded_avg_last5":     hrf["conceded"],
            "away_goals_conceded_avg_last5":     arf["conceded"],
            "home_xg_avg_last5":                 hrf["xg"],
            "away_xg_avg_last5":                 arf["xg"],
            "home_ppda_avg_last5":               hrf["ppda"],
            "away_ppda_avg_last5":               arf["ppda"],
            "home_deep_completions_avg_last5":   hrf["deep"],
            "away_deep_completions_avg_last5":   arf["deep"],
            "home_points_last5":                 hpt,
            "away_points_last5":                 apt,
            "home_team_home_wins_last5":          hvf["wins"],
            "home_team_home_draws_last5":         hvf["draws"],
            "home_team_home_losses_last5":        hvf["losses"],
            "away_team_away_wins_last5":          avf["wins"],
            "away_team_away_draws_last5":         avf["draws"],
            "away_team_away_losses_last5":        avf["losses"],
            "points_diff_last5":    hpt - apt,
            "goal_diff_avg5":       hrf["goals"]    - arf["goals"],
            "xg_diff_avg5":         hrf["xg"]       - arf["xg"],
            "x_defense_diff":       arf["conceded"] - hrf["conceded"],
            "ppda_diff_avg5":       hrf["ppda"]     - arf["ppda"],
            "deep_comp_diff_avg5":  hrf["deep"]     - arf["deep"],
            "venue_wins_diff":      hvf["wins"]     - avf["wins"],
            "home_venue_advantage": hvf["win_rate"] - avf["win_rate"],
            "home_advantage":       1,
        }

        X          = pd.DataFrame([row])[self.feature_cols]
        pred_enc   = self.model.predict(X)[0]
        pred_proba = self.model.predict_proba(X)[0]
        pred_label = self.le.inverse_transform([pred_enc])[0]

        proba_dict = {
            cls: round(float(p) * 100, 2)
            for cls, p in zip(self.le.classes_, pred_proba)
        }

        return {
            "home_team":        home_team,
            "away_team":        away_team,
            "match_date":       match_date,
            "predicted_result": pred_label,
            "probabilities":    proba_dict,
            "feature_row":      X,
            "_hrf": hrf, "_arf": arf,
            "_hvf": hvf, "_avf": avf,
            "_hpt": hpt, "_apt": apt,
        }

    def predict_pretty(self, home_team: str, away_team: str,
                       match_date: str):
        result = self.predict(home_team, away_team, match_date)
        proba  = result["probabilities"]
        hvf, avf = result["_hvf"], result["_avf"]
        hpt, apt = result["_hpt"],  result["_apt"]

        h_venue = f"W{hvf['wins']} D{hvf['draws']} L{hvf['losses']}"
        a_venue = f"W{avf['wins']} D{avf['draws']} L{avf['losses']}"
        ht = home_team[:24]
        at = away_team[:24]

        print("=" * 58)
        print(f"  {home_team}  vs  {away_team}")
        print(f"  Date: {match_date}")
        print("=" * 58)
        print(f"  {'Team':<26}  {'Pts last 5':>10}  {'Venue (last 5)'}")
        print(f"  {'-'*55}")
        print(f"  {ht:<26}  {hpt:>10.0f}  At home : {h_venue}")
        print(f"  {at:<26}  {apt:>10.0f}  Away    : {a_venue}")
        print("-" * 58)
        print(f"  Predicted result : {result['predicted_result']}")
        print("-" * 58)
        print("  Probabilities:")
        for label in ["Win", "Draw", "Lose"]:
            bar_len = int(proba[label] / 2.5)
            bar     = "█" * bar_len
            print(f"    {label:<5}  {bar:<40}  {proba[label]:5.1f}%")
        print("=" * 58)
        return result


# ============================================================
# CELL 20 — Instantiate predictor
# ============================================================

predictor = EPLPredictor(
    historical_df=raw_df,
    model=model,
    label_encoder=le,
    feature_cols=INPUT_FEATURES,
)
print("Predictor ready.")
print(f"Teams available: {sorted(raw_df['home_team'].unique())}")


# ============================================================
# CELL 21 — Predict: Manchester City vs Arsenal
# ============================================================

result_1 = predictor.predict_pretty(
    home_team  = "Manchester City",
    away_team  = "Arsenal",
    match_date = "2026-03-20",
)


# ============================================================
# CELL 22 — Predict: Liverpool vs Chelsea
# ============================================================

result_2 = predictor.predict_pretty(
    home_team  = "Liverpool",
    away_team  = "Chelsea",
    match_date = "2026-03-20",
)


# ============================================================
# CELL 23 — Predict any match (template)
# ============================================================

HOME_TEAM  = "Manchester United"
AWAY_TEAM  = "Tottenham"
MATCH_DATE = "2026-04-05"

result_custom = predictor.predict_pretty(HOME_TEAM, AWAY_TEAM, MATCH_DATE)


# ============================================================
# CELL 24 — Feature importance bar chart
# ============================================================

feat_df = (
    pd.Series(model.feature_importances_, index=INPUT_FEATURES)
    .sort_values(ascending=False)
    .reset_index()
)
feat_df.columns = ["Feature", "Importance"]

plt.figure(figsize=(10, 9))
colors = ["#1d4ed8" if i < 5 else "#93c5fd" for i in range(len(feat_df))]
plt.barh(feat_df["Feature"][::-1], feat_df["Importance"][::-1], color=colors[::-1])
plt.xlabel("Importance Score")
plt.title("Feature Importances — EPL Match Predictor (AdaBoost)", fontweight="bold")
plt.tight_layout()
plt.savefig("feature_importances.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved as feature_importances.png")


# ============================================================
# CELL 25 — Prediction probability bar chart
# ============================================================

def plot_prediction(result: dict):
    labels = list(result["probabilities"].keys())
    values = list(result["probabilities"].values())
    colors = ["#16a34a" if l == "Win"
              else "#ca8a04" if l == "Draw"
              else "#dc2626" for l in labels]

    plt.figure(figsize=(6, 4))
    bars = plt.bar(labels, values, color=colors, edgecolor="white", linewidth=1.5)
    plt.ylim(0, 100)
    plt.ylabel("Probability (%)")
    plt.title(
        f"{result['home_team']} vs {result['away_team']}\n"
        f"Predicted: {result['predicted_result']}",
        fontweight="bold"
    )
    for bar, val in zip(bars, values):
        plt.text(bar.get_x() + bar.get_width() / 2,
                 bar.get_height() + 1.5,
                 f"{val:.1f}%", ha="center", fontsize=11, fontweight="bold")
    plt.tight_layout()
    plt.savefig(
        f"pred_{result['home_team'].replace(' ','_')}"
        f"_vs_{result['away_team'].replace(' ','_')}.png",
        dpi=150, bbox_inches="tight"
    )
    plt.show()

plot_prediction(result_1)
plot_prediction(result_2)


# ============================================================
# CELL 26 — Save model
# ============================================================

import pickle

with open("epl_adaboost_model.pkl", "wb") as f:
    pickle.dump(model, f)
with open("epl_label_encoder.pkl", "wb") as f:
    pickle.dump(le, f)

print("Model saved  : epl_adaboost_model.pkl")
print("Encoder saved: epl_label_encoder.pkl")


# ============================================================
# CELL 27 — Load saved model and predict (standalone)
# ============================================================

with open("epl_adaboost_model.pkl", "rb") as f:
    loaded_model = pickle.load(f)
with open("epl_label_encoder.pkl", "rb") as f:
    loaded_le = pickle.load(f)

loaded_predictor = EPLPredictor(
    historical_df=raw_df,
    model=loaded_model,
    label_encoder=loaded_le,
    feature_cols=INPUT_FEATURES,
)

loaded_predictor.predict_pretty(
    home_team  = "Manchester City",
    away_team  = "Liverpool",
    match_date = "2026-04-12",
)
```
